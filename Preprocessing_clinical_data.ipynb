### Handling earlier_therapy

```python
print(clinical_data['earlier_therapy'].unique())
```

```python
import re
```

```python
### Step 1: Extract atomic therapies ###
def extract_atoms(therapy_str: str) -> list:
    """Splits therapy string into atomic components."""
    if therapy_str.lower().strip() in ['none', 'unknown', '', None]:
        return []
    return re.split(r'\s*\+\s*', therapy_str.lower().strip())
```

```python
### Step 2: Build vocabulary ###
def build_therapy_vocab(therapy_series: pd.Series) -> list:
    """Build vocabulary of atomic therapies."""
    atom_set = set()
    for entry in therapy_series:
        atom_set.update(extract_atoms(entry))
    return sorted(atom_set)

therapy_vocab = build_therapy_vocab(clinical_data['earlier_therapy'])
print("Therapy Vocabulary:", therapy_vocab)
```

```python
### Step 3: Multi-hot encode each entry ###
def multi_hot_encode(therapy_str: str, vocab: list) -> list:
    """Multi-hot encodes a single therapy string."""
    atoms = extract_atoms(therapy_str)
    return [1 if v in atoms else 0 for v in vocab]
```

```python
# Apply encoding
clinical_data['therapy_vector'] = clinical_data['earlier_therapy'].apply(
    lambda x: multi_hot_encode(x, therapy_vocab)
)
```

Convert therapy vector from list to numerical list
